#  1. Data Wrangling & Preprocessing

```python
import pandas as pd
import numpy as np

# Create a copy of the original dataset
df_cleaned = df.copy()

# Step 1: Positional columns to transform
positional_cols = ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw',
                   'lam', 'cam', 'ram', 'lm', 'lcm', 'cm', 'rcm', 'rm',
                   'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb',
                   'rcb', 'rb']

# Step 1.1: Function to extract bonus % score
def extract_base_bonus(val):
    if isinstance(val, str) and '+' in val:
        base, bonus = val.split('+')
        return float(base), float(bonus)
    try:
        return float(val), 0.0
    except:
        return np.nan, 0.0

def compute_bonus_percent(base, bonus):
    if pd.isna(base) or base == 0:
        return np.nan
    return round((bonus / base) * 100, 2)

# Step 1.2: Create new score columns and drop original ones
for col in positional_cols:
    if col in df_cleaned.columns:
        extracted = df_cleaned[col].apply(extract_base_bonus)
        df_cleaned[f'{col}_score'] = [compute_bonus_percent(base, bonus) for base, bonus in extracted]
        df_cleaned.drop(columns=[col], inplace=True)

# Step 2: Drop irrelevant or unneeded columns
columns_to_drop = ['sofifa_id', 'player_url', 'long_name', 'dob']
df_cleaned.drop(columns=columns_to_drop, inplace=True)

# Step 3: Identify columns with more than 90% missing values
high_missing_cols = df_cleaned.columns[df_cleaned.isnull().mean() > 0.90].tolist()

# âœ… Output
print("âœ… Positional performance bonus columns created. Sample output:")
display(df_cleaned[[col + '_score' for col in positional_cols]].head())

print(f"âš ï¸ Columns with more than 90% missing values:\n{high_missing_cols}")
```

## âœ… Code: Parse + Impute value_eur and wage_eur in df_cleaned

```python
def parse_currency(val):
    if pd.isnull(val): return np.nan
    if isinstance(val, (int, float)): return val
    val = str(val).replace('â‚¬', '').strip().upper()
    if 'M' in val:
        return float(val.replace('M', '')) * 1_000_000
    elif 'K' in val:
        return float(val.replace('K', '')) * 1_000
    try:
        return float(val)
    except:
        return np.nan

for col in ['value_eur', 'wage_eur']:
    df_cleaned[col] = df_cleaned[col].apply(parse_currency)

    # Replace 0 values with the median of non-zero values
    median_val = df_cleaned.loc[df_cleaned[col] != 0, col].median()
    df_cleaned[col] = df_cleaned[col].replace(0, median_val)

    print(f"âœ… '{col}' parsed and 0s replaced.")
    print(f"   {col} median used: â‚¬{int(median_val):,}")
```

#ðŸ’° Log-Transformed Financial Features
##ðŸ§¾ Description:
To reduce skewness and handle the wide range of monetary values in the FIFA 20 dataset, we applied natural logarithmic transformation (log1p, i.e., log(x + 1)) to the following financial features:

#value_eur â†’ log_value_eur
Represents the market value of the player in euros.
The transformed column captures relative value differences more effectively by compressing large ranges and making the distribution more normal-like.

#wage_eur â†’ log_wage_eur
Represents the player's weekly wage in euros.
Log transformation reduces the extreme imbalance caused by a few highly paid players.

#release_clause_eur â†’ log_release_clause_eur
Represents the release clause amount (minimum transfer fee) in euros.
The transformation helps in dealing with large variations and zero values.

#âœ… Why it's useful:
Handles outliers and skewed distributions.

Improves performance of regression and distance-based models.

Retains interpretability via the log scale.
